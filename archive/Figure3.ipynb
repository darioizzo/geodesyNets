# Loading and visualizing the ground truth asteroid (a point cloud)

```python
with open("mascons/Hollow_nu.pk", "rb") as file:
    hollow_p, hollow_m, _ = pk.load(file)
with open("mascons/Torus.pk", "rb") as file:
    torus_p, torus_m, _ = pk.load(file)
with open("mascons/Bennu_nu.pk", "rb") as file:
    bennu_p, bennu_m, _ = pk.load(file)
with open("mascons/Eros.pk", "rb") as file:
    eros_p, eros_m, _ = pk.load(file)
with open("mascons/Churyumov-Gerasimenko.pk", "rb") as file:
    p67_p, p67_m, _ = pk.load(file)
with open("mascons/Itokawa_nu.pk", "rb") as file:
    ito_p, ito_m, _ = pk.load(file)
    


def plot_mascon(ax, p, alpha = 0.05, s=20, D=0.5, elev=45, azim=45):
    ax.scatter(p[:,0], p[:,1], p[:,2], alpha=alpha, s=s, c='k')
    ax.view_init(elev=elev, azim=azim)
    ax.set_xlim([-D, D])
    ax.set_ylim([-D, D])
    ax.set_zlim([-D, D])
    ax.set_axis_off()
    ax.quiver([0],[0],[0], [1], [0], [0], length=0.2, color='red')
    ax.quiver([0],[0],[0], [0], [0], [1], length=0.2, color='green')
    ax.quiver([0],[0],[0], [0], [1], [0], length=0.2, color='blue')
```

```python
fig = plt.figure(figsize=(9.8,9.8))
#Hollow
ax = fig.add_subplot(231, projection='3d')
plot_mascon(ax, hollow_p, s = hollow_m * 150000, D=0.6)
ax.set_title('Hollow')



#Torus
ax = fig.add_subplot(232, projection='3d')
plot_mascon(ax, torus_p, s = torus_m * 20000, D=0.51)
ax.set_title('Torus')

#Eros
ax = fig.add_subplot(233, projection='3d')
plot_mascon(ax, eros_p, s = eros_m * 10000, D=0.51)
ax.set_title('Eros')


#67p
ax = fig.add_subplot(234, projection='3d')
plot_mascon(ax, p67_p, s = p67_m * 10000, D = 0.50, azim=150)
ax.set_title('Churyumov-Gerasimenko')

#Itokawa
ax = fig.add_subplot(235, projection='3d')
plot_mascon(ax, ito_p, s = ito_m * 10000, D = 0.51, azim=-61, elev = -170)
ax.set_title('Itokawa')

#Bennu
ax = fig.add_subplot(236, projection='3d')
plot_mascon(ax, bennu_p, s = bennu_m * 20000, D = 0.51, azim=300, elev = 10)
ax.set_title('Bennu')

plt.tight_layout()
plt.savefig("Fig2.png", dp=600)
```
